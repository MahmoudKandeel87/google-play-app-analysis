```python
import numpy as np
import pandas as pd
import seaborn as sns


sns.set(rc={'figure.figsize':[9,9]},font_scale = 1.4 )
```

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
s = df["Size"].unique()
```

```python
for j in s:
    if ('k' not in j) and ('M' not in j):
        print(j)
```

```python
idx = df[ df['Size']. isin( ["Varies with device","1,000+"])].index
```

```python
idx
```

```python
df.drop(idx,axis = 0 , inplace = True)
```

```python
df["Size"].unique()
```

```python
def size_k_num(s):
    if s.endswith('M'):
        m = float(s[:-1])
        return m * 1024
    elif s.endswith('k'):
        return float(s[:-1])
    
    
```

```python
size_k_num("861k")
```

```python
size_k_num('65M')
```

```python
df["Size"] = df["Size"].apply(size_k_num)
```

```python
df
```

```python
def is_ver4(v):
    try:
        return v.startswith("4")
    except:
        return False
```

```python
is_ver4("2.2 and up")
```

```python
is_ver4("4.0.3 and up")
```

```python
df['is_var4'] = df['Android Ver'].apply(is_ver4) 
```

```python
df
```

```python
df['Reviews'] = df['Reviews'].apply(int)
```

```python
df.info()
```

```python
df['Price'].unique()
```

```python
def get_float_price(p):
    try:
        return float(p[1:])
    except:
        return 0.0
```

```python
get_float_price("0")
```

```python
df["Price"] = df["Price"].apply(get_float_price)
```

```python
df.info()
```
